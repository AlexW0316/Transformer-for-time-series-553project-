```python
#import pandas as pd

#from google.colab import drive
#drive.mount('/content/drive')

import os
directory = "/root/.cache/kagglehub/datasets/saketk511/2019-2024-us-stock-market-data/versions/1"
print(os.listdir(directory))

import pandas as pd

csv_path = "/root/.cache/kagglehub/datasets/saketk511/2019-2024-us-stock-market-data/versions/1/"
df = pd.read_csv(os.path.join(csv_path, "Stock Market Dataset.csv") )

# Drop the first column
df = df.iloc[:, 1:]  # Select all columns except the first one

# Target column is the apple stock price
target_column = 'Gold_Price'

# Select all columns except the target and date for features
features = [col for col in df.columns if col not in ['Date', target_column]]

# Include the target column for supervised learning
selected_columns = ['Date', target_column] + features
data = df[selected_columns]

# Remove commas and convert all numeric columns to float
for col in data.columns[1:]:  # Skip the 'Date' column
    data[col] = data[col].astype(str).str.replace(',', '').astype(float)

# Convert date column to datetime
data.loc[:, 'date'] = pd.to_datetime(data['Date'],  format="%d-%m-%Y")

# Sort by date
data = data.sort_values(by='date')

# Reset the index
data.reset_index(drop=True, inplace=True)

# Fill or drop missing values as needed
data.ffill(inplace=True)  # Forward fill

# Move "date" to the first column
cols = data.columns.tolist()
data = data[[cols[-1]] + cols[:-1]]

# Remove the "Date" column
data = data.drop(columns=['Date'])

result = pd.concat([data.iloc[:, :8], data.iloc[:, -1]], axis=1)

#from sklearn.preprocessing import StandardScaler

# Initialize a MinMaxScaler
#scaler = StandardScaler()

#data[data.columns[1:]] = scaler.fit_transform(data[data.columns[1:]])
```
